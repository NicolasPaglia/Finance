```python
import yfinance as yf
import pandas as pd

# Diccionario de métricas categorizadas por segmento
categories = {
    'Valuación': ['P/E', 'PEG 5 años', 'Crecimiento Implícito (P/E sobre PEG)', 'P/S', 'P/B', 'EV/Revenue', 'EV/EBITDA'],
    'Dividendos': ['Rendimiento del Dividendo (Yield)', 'Razón de Pago de Dividendos (Payout)'],
    'Rentabilidad': ['ROA', 'ROE', 'Margen de Utilidad'],
    'Solvencia': ['Deuda a Patrimonio (D/E)', 'Cobertura de Intereses'],
    'Liquidez': ['Razón Corriente', 'Prueba Ácida'],
    'Riesgo': ['Beta (5 años, mensual)']
}



# Expandir categorías en un DataFrame
data = {'Segmento': [], 'KPI': []}
for category, metrics in categories.items():
    data['Segmento'].extend([category] * len(metrics))
    data['KPI'].extend(metrics)

# Crear el DataFrame inicial
financial_data = pd.DataFrame(data)

# Extraer métricas para cada ticker
def get_financial_metrics(ticker):
    stock = yf.Ticker(ticker)
    info = stock.info

    pe = info.get('trailingPE')
    peg = info.get('trailingPegRatio')

    crecimiento_implicito = pe / peg if pe and peg else None

    return [
        # Valuación
        pe,
        peg,
        crecimiento_implicito,
        info.get('priceToSalesTrailing12Months'),
        info.get('priceToBook'),
        info.get('enterpriseToRevenue'),
        info.get('enterpriseToEbitda'),
        # Dividendos
        info.get('dividendYield') * 100 if info.get('dividendYield') else None,
        info.get('payoutRatio'),
        # Rentabilidad
        info.get('returnOnAssets'),
        info.get('returnOnEquity'),
        info.get('profitMargins'),
        # Solvencia
        info.get('debtToEquity'),
        info.get('interestCoverage'),
        # Liquidez
        info.get('currentRatio'),
        info.get('quickRatio'),
        # Riesgo
        info.get('beta')
    ]

# Agregar los datos de cada ticker al DataFrame
for ticker in tickers:
    financial_data[ticker] = get_financial_metrics(ticker)

# Ajustar el formato del DataFrame para que se visualicen las categorías
financial_data = financial_data[['Segmento', 'KPI'] + tickers]


# Extraer los valores típicos de la industria
industry_metrics = []
for segmento, metrics in industria_kpi[selected_industry].items():
    for kpi, value in metrics.items():
        industry_metrics.append({'Segmento': segmento, 'KPI': kpi, 'Industria_Típico': value})

# Convertir a DataFrame
industry_data = pd.DataFrame(industry_metrics)

# Hacer un merge con financial_data
merged_data = pd.merge(financial_data, industry_data, on=['Segmento', 'KPI'], how='left')

# # Mostrar el resultado
# # Crear un DataFrame para las descripciones
# description_data = pd.DataFrame(list(descriptions.items()), columns=['KPI', 'Descripción'])

# # Hacer un merge con el DataFrame existente
# merged_data_with_descriptions = pd.merge(merged_data, description_data, on='KPI', how='left')

# # Mostrar el resultado
# merged_data_with_descriptions
```

### Teoría de Portafolio (Markowitz)

```python
import yfinance as yf
from pypfopt import EfficientFrontier, risk_models, expected_returns, plotting
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Obtener Datos de YFinance
datos = yf.download(tickers, start = "2020-01-01", end = "2024-01-01", interval = '1mo')['Adj Close'] # Datos Mensuales

# datos.info()
datos_limpio = datos.dropna(axis=1, how = "any")
# datos_limpio

# Rendimientos
mu = expected_returns.mean_historical_return(datos_limpio, frequency = 12)

# Matriz varianza- covarianza
S = risk_models.sample_cov(datos_limpio, frequency = 12)


tasa_libre_riesgo = 0.175

ef = EfficientFrontier(mu, S)


pesos_optimizados = ef.max_sharpe(risk_free_rate = tasa_libre_riesgo)



rendimiento, volatilidad, razon_sharpe= ef.portfolio_performance(verbose = True)

pesos_optimizados
```

#### Frontera Eficiente

```python
ef = EfficientFrontier(mu, S)

# Función para calcular la CML
def capital_market_line(tasa_libre_riesgo):
    slope = (rendimiento - tasa_libre_riesgo) / (volatilidad - 0)  # Pendiente de la CML
    return lambda x: tasa_libre_riesgo + slope * x

# Graficar la frontera eficiente
fig, ejes = plt.subplots(figsize=(10, 5), dpi=300)
plotting.plot_efficient_frontier(ef, ax=ejes, show_assets=True)

# Marcar el portafolio óptimo
ejes.scatter(volatilidad, rendimiento, marker='*', color='red', s=100, label='Portafolio Óptimo')

for activo, peso in pesos_optimizados.items():
    # Calcula la volatilidad y el retorno esperado del activo
    volatilidad_activo = S.loc[activo, activo] ** 0.5
    retorno_esperado_activo = mu[activo]

    # Marca el activo en el gráfico
    ejes.annotate(activo, 
                  (volatilidad_activo, retorno_esperado_activo), 
                  color='red' if peso > 0 else 'lightgray', 
                  fontsize=8)  # Cambio de color a gris claro y reducción del tamaño de la fuente

# Graficar la CML
cml = capital_market_line(tasa_libre_riesgo)
x = np.linspace(0, volatilidad * 2, 100)  # Ajuste en el rango de x para evitar el error
ejes.plot(x, cml(x), color="green", linestyle="--", label="Línea de Mercado de Capitales")

# Detalles del gráfico
ejes.set_title('Frontera Eficiente, Portafolio Óptimo y Línea de Mercado de Capitales')
ejes.set_xlabel('Volatilidad')
ejes.set_ylabel('Rendimiento Esperado')
ejes.legend()
plt.show()
```

### Simulación de Montecarlo


```python
import numpy as np
import matplotlib.pyplot as plt

# Datos iniciales
S0 = 100  # Precio actual
mu = 1.030197  # Rendimiento promedio anual
sigma = 0.567267  # Volatilidad anual
T = 1  # Horizonte de tiempo en años
n_simulaciones = 1000  # Número de simulaciones
n_pasos = 252  # Días en un año de trading

# Simulación
dt = T / n_pasos
precios = np.zeros((n_pasos, n_simulaciones))
precios[0] = S0

for t in range(1, n_pasos):
    Z = np.random.standard_normal(n_simulaciones)  # Aleatorio normal
    precios[t] = precios[t - 1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)

# Visualizar resultados
plt.plot(precios[:, :100])  # Mostrar 10 trayectorias
plt.title("Simulaciones Monte Carlo de Precios de la Acción")
plt.xlabel("Días")
plt.ylabel("Precio")
plt.show()
```

```python
# Precios finales al final del período
precios_finales = precios[-1]

# Histograma
plt.hist(precios_finales, bins=50, color='skyblue', edgecolor='black')
plt.title("Distribución de Precios Finales - Simulación Monte Carlo")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```
